```python
num = [9]
den1 = [1, 9, 9]
den2 = [1, 2, 9]
den3 = [1, 0, 9]
den4 = [1, 6, 9]
den5 = [1, 0, -9]

sis1 = ct.tf(num, den1)
sis2 = ct.tf(num, den2)
sis3 = ct.tf(num, den3)
sis4 = ct.tf(num, den4)
sis5 = ct.tf(num, den5)
```

```python
T = np.linspace(0, 10, 1000)

resp1 = ct.step_response(sis1, T = T)
resp2 = ct.step_response(sis2, T = T)
resp3 = ct.step_response(sis3, T = T)
resp4 = ct.step_response(sis4, T = T)
resp5 = ct.step_response(sis5, T = T)
```

## Modelagem de um sistema el√©trico

```python
num = [1]
den = [2, 4, 2]

sis = ct.tf(num, den)
```

### Resposta do sistema sob degrau unit√°rio

```python
T = np.linspace(0, 10, 1000)

resp6 = ct.step_response(sis, T = T)
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(10,5), dpi = 80)

ax1.plot(resp6.time, resp6.inputs, label = 'Degrau', color = 'blue')
ax2 = ax1.twinx()
ax2.plot(resp6.time, resp6.states, label = 'Resposta', color = 'red')

ax1.set(title = 'Simula√ß√£o da Resposta do Circuito sob Degrau Unit√°rio', xlabel = 'Tempo [s]', ylabel = '$V_L [V]$')

ax2.set_ylabel('$V_i [A]$')

ax1.grid()
ax2.grid()

ax1.legend(loc = 'lower center')
ax2.legend()

plt.show()
```

### Resposta do sistema sob rampa unit√°rio

```python
T = np.linspace(0, 10, 1000)
U = np.ones((1000))*T/max(T)
resp7 = ct.forced_response(sis, T = T, U = U)
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(10,5), dpi = 80)

ax1.plot(resp7.time, U, label = 'Rampa', color = 'blue')
ax2 = ax1.twinx()
ax2.plot(resp7.time, resp7.states, label = 'Resposta', color = 'red')

ax1.set(title = 'Simula√ß√£o da Resposta do Circuito sob Rampa Unit√°rio', xlabel = 'Tempo [s]', ylabel = '$V_L [V]$')

ax2.set_ylabel('$V_i [A]$')

ax1.grid()
ax2.grid()

ax1.legend(loc = (0.85, 0.55))
ax2.legend(loc = (0.85, 0.45))

plt.show()
```
