```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
Gn = ct.tf(820, [1, 30, 200])
Gc = ct.tf([1, 0.05], [1, 0.005])
G = ct.series(Gn, Gc)

sisn = ct.feedback(Gn,1)
sis = ct.feedback(G,1)
```

```python
ct.root_locus(Gn)
plt.plot()
```

```python
ct.root_locus(Gc)
plt.plot()
```

```python
ct.root_locus(G)
plt.plot()
```

```python
res_n = ct.step_response(sisn, T = np.linspace(0,100,10000))
res = ct.step_response(sis, T = np.linspace(0,100,10000))
```

```python
fig, [ax1, ax2] = plt.subplots(2,1, figsize=(10,10), dpi=200)

ax1.plot(res_n.time, res_n.outputs, label='No Compensado')
ax1.plot(res.time, res.outputs, label='Compensado')
ax1.grid()
ax1.legend()
ax1.set(title='Comparao da resposta permanente \n no tempo do sitema compensado e no compensado', 
       xlabel='Tempo [s]', ylabel='x [-]')

ax2.plot(res_n.time, res_n.outputs, label='No Compensado')
ax2.plot(res.time, res.outputs, label='Compensado')
ax2.grid()
ax2.legend()
ax2.set_xlim([0, 5])
ax2.set(title='Comparao da resposta transiente \n no tempo do sitema compensado e no compensado', 
       xlabel='Tempo [s]', ylabel='x [-]')
fig.tight_layout()

plt.savefig('PT8-sis_comp.jpg')

plt.show()
```
